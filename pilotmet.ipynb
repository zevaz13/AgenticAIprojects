```python
import pygame
import time

# -----------------------------
# User parameters
# -----------------------------
flicker_hz = 10
screen_width_cm = 34.5   # change to your monitor width in cm
screen_res_x = 1920      # change to your monitor resolution
square_size_cm = 1.0

# Known comparison color
known_color = (223, 112, 0)

# -----------------------------
# Derived parameters
# -----------------------------
pixels_per_cm = screen_res_x / screen_width_cm
square_size_px = int(square_size_cm * pixels_per_cm)

period = 1.0 / flicker_hz
half_period = period / 2

# -----------------------------
# Initialize pygame
# -----------------------------
pygame.init()
screen = pygame.display.set_mode((800, 600))
clock = pygame.time.Clock()

font = pygame.font.SysFont("Arial", 24)

# Starting mixture color
red = 128
green = 128
blue = 0

running = True
last_switch = time.time()

# -----------------------------
# Slow update control
# -----------------------------
last_update = 0
update_delay = 40   # ms between increments (try 30–80)

while running:
    now = time.time()
    now_ms = pygame.time.get_ticks()

    # Handle quit events
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # -----------------------------
    # Continuous key controls (slow)
    # -----------------------------
    keys = pygame.key.get_pressed()

    if now_ms - last_update > update_delay:

        # Red controls
        if keys[pygame.K_t]:
            red = min(255, red + 1)
        if keys[pygame.K_r]:
            red = max(0, red - 1)

        # Green controls
        if keys[pygame.K_s]:
            green = min(255, green + 1)
        if keys[pygame.K_a]:
            green = max(0, green - 1)

        last_update = now_ms

    # -----------------------------
    # Flicker logic
    # -----------------------------
    if now - last_switch >= half_period:
        show_mixture = not locals().get("show_mixture", True)
        last_switch = now

    # -----------------------------
    # Drawing
    # -----------------------------
    screen.fill((255, 255, 255))

    # Choose which color to show
    if show_mixture:
        color = (red, green, blue)
    else:
        color = known_color

    # Draw the square
    pygame.draw.rect(
        screen,
        color,
        pygame.Rect(100, 100, square_size_px, square_size_px)
    )

    # Draw RGB text
    text_surface = font.render(
        f"Mixture RGB = ({red}, {green}, {blue})", True, (0, 0, 0)
    )
    screen.blit(text_surface, (100, 200))

    pygame.display.flip()
    clock.tick(120)

pygame.quit()
```
